1  # Queues
2
3  A queue is an ordered collection of elements where items are added and removved from **opposite** ends, commonly referred to as the "front" and "rear" of the queue.  The ordering principle is referred to as FIFO (First In, First Out) or "first come first come first served.

```python
# Challenge #1

stack = []

stack.append("A")
stack.append("B")
stack.append("C")

top = stack.pop()

print(f"Current stack: {stack}")
print(f"Value removed: {top}")

popped_value = stack.pop()
print(f"Next value removed: {popped_value}")

stack.append("D")

last_pop = stack.pop()
print(f"Popped after adding D: {last_pop}")

print(f"Final stack state: {stack}")
```

```python
# Challenge #4

stack = []
word = "Hello"


for char in word:
    stack.append(char)


inverted_word = ""
while len(stack) > 0:
    inverted_word += stack.pop()

print(inverted_word)
```

```python
# Challenge #5

from collections import deque

# 1. Create an empty queue
queue = deque()

# 2. Enqueue the values 10, 20, and 30
queue.append(10)
queue.append(20)
queue.append(30)

# 3. Print the queue
print(f"Current Queue: {list(queue)}")
```

```python
#Challenge #6
from collections import deque

# 1. Create a queue with the values 5, 6, 7, 8
queue = deque([5, 6, 7, 8])

# 2. Dequeue one item (removes the first item added)
removed_item = queue.popleft()

# 3. Print the value removed and the final queue
print(f"Value removed: {removed_item}")
print(f"Final queue: {list(queue)}")
```
