```python
nums = [1, 4, 51, 200, 14, 8]

def find_largest(nums):
    if not nums:
        return "no values"
    
    largest = nums[0]

    for n in nums[1:]:
        if n >largest:
            largest = n

    return largest

print(find_largest(nums))
```
